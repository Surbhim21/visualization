```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)

# make a figure and within that figure we have axes plot
fig, ax = plt.subplots(figsize=(5, 3))
# ax.scatter(x,y) #how we want our data to look
# ax.bar(x,y)
# ax.barh(x,y)
# ax.plot(x,y)
ax.hist(y)
```

```python
font1 = {'family': 'sans-serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,
           y,
           marker='*',  # indicators # D = diamonds, > for arrows
           color='indigo')
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*',
        color='#7425B9',
        linestyle='--',
        linewidth=2,
        markersize=12,
        markeredgecolor='#FA9359',
        markerfacecolor='#000000')
# ax.grid() - gird line on both x and y axis
ax.grid(axis='y',
        linestyle=':',
        linewidth='2',
        color='#000000')
```

```python
df = pd.DataFrame({'group': list(map(chr, range(65, 85))),
                  'values': np.random.uniform(size=20)})

# Reorder it following the values:
ordered_df = df.sort_values(by='values')
my_range = range(1, len(df.index)+1)

# Make the plot
plt.stem(ordered_df['values'])
plt.xticks(my_range, ordered_df['group'])
```
