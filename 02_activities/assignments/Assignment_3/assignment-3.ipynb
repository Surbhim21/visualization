```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

np.random.seed(100)
```

```python
df = pd.read_csv("Assignment_3/Traffic Signal - 4326.csv")
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df['ACTIVATIONDATE'] = pd.to_datetime(df['ACTIVATIONDATE'], errors='coerce')
df['Activation_Year'] = df['ACTIVATIONDATE'].dt.year

fig_hist = px.histogram(df, x='Activation_Year',

                        color='CONTROL_MODE',  # categorical variable for color

                        nbins=30,

                        title='Number of Traffic Signals Installed Per Year by Control Mode')

fig_hist.show()
```
