```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL  # install as pillow if needed
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# plot both variables on the same axis

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='Person 1')  # to put
ax.plot(x, y2, label='Person 2')
ax.legend(loc='lower right',
          frameon=True,
          fontsize=12,
          ncol=2,
          shadow=True
          )
fig.show()
```

```python
# to put the legends outside of the box
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
ax.legend(loc='upper left',
          bbox_to_anchor=(1, 1))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc="lower right")
ax.text(10, 95, 'This value is important!',  # 10,95 are the coordinates and it makes the ceter towards it
        ha='center',
        color='red',
        size=20)
```

```python
# position text annotations

fig, ax = plt.subplots()  # simple empty set of axes
ax.axis([0, 10, 0, 10])

ax.text(1, 5, ".Data: (1,5)", transform=ax.transData)
ax.text(0.5, 0.1, ".Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ".Figures: (0.2, 0.2)", transform=fig.transFigure)
```

```python
# remove tick marks and labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label=' Person 2')
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())  # removes both ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label=' Person 2')
ax.legend(loc='lower right')

# ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #to limit the number of labels
# ax.xaxis.set_major_locator(plt.MultipleLocator(10)) #intervals bw labels

plt.xticks(rotation=45, ha='right', fontsize=12, font='serif')

# activity
font1 = {'family': 'serif', 'color': 'indigo'}
# Practice: font2 = {'family': 'calibri', 'color': 'black'}
plt.xlabel('Shiny Nex X Axis!', fontsize=18, fontdict=font1)
# Practice: plt.ylabel ('Shiny New y-axis!', fontsize = 18, fontdict = font2)
```

```python
np.random.seed(200)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Q1", "Q2", "Q3", "Q4"])
y2 = np.array([110, 180, 240, 99])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7, 3))

ax1.scatter(x1, y1,
            marker="*",
            color='black')
ax2.bar(x2, y2,
        color='orange')

ax1.set_title("Scatterplot", fontdict=font1)
ax1.set_xlabel("Customers")
ax2.set_title("Sales", fontdict=font1)
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFD700']
ax2.bar(x2, y2, color=colors)
```
